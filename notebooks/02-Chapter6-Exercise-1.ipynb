```python
from pathlib import Path
import pandas as pd
import numpy as np
import altair as alt
```

```python
src_file = Path.cwd()/'..'/'code'/'data'/'raw'/'EPA_fuel_economy_summary.csv'
```

```python
df = pd.read_csv(src_file)
df.head()
```

```python
alt.data_transformers.enable('vegafusion')
```

```python
chart = alt.Chart(df).mark_circle().encode(x='displ',y='fuelCost08')
```

```python
chart.display()
```

```python
chart = alt.Chart(df).mark_point().encode(x='displ',y='fuelCost08')
chart.display()
```
