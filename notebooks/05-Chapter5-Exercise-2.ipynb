```python
from pathlib import Path
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
src_file = Path.cwd()/'..'/'code'/'data'/'raw'/'EPA_fuel_economy_summary.csv'
df = pd.read_csv(src_file)
```

```python
df.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.displot(data=df, x='fuelCost08', hue='year', aspect=2.5)
```

```python
sns.displot(data=df, x='fuelCost08', hue='date_range', aspect=2.5)
```

```python
sns.displot(data=df, x='fuelCost08', col='date_range')
```

```python
sns.displot(data=df, x='fuelCost08', row='date_range', kind='kde')
```

Thanks for [chatgpt](https://chatgpt.com/share/67e84a5c-2bd0-8001-8b71-429f194b53fa) for the following chart

```python
sns.kdeplot(data=df, x='fuelCost08', hue='date_range')
plt.title('Distribution of fuelCost by Date Range')
plt.show()
```

```python
sns.relplot(data=df, x='cylinders', y='comb08', kind='scatter')
```

```python
sns.relplot(data=df, x='cylinders', y='comb08', kind='scatter', hue='drive')
```

```python
sns.relplot(data=df, x='cylinders', y='comb08', kind='scatter', hue='drive', col='date_range')
```

```python
sns.relplot(data=df, x='cylinders', y='comb08', kind='line', hue='drive', col='date_range')
```

```python
sns.catplot(data=df,
            x='date_range',
            y='comb08',
            kind='box',
            showfliers=False);
```

```python
sns.catplot(data=df,
            x='date_range',
            y='comb08',
            kind='boxen',
            showfliers=False);
```

```python
sns.catplot(data=df,
            x='date_range',
            y='comb08',
            kind='box',
            showfliers=False,
           col='fuel_type_summary');
```

```python
sns.catplot(data=df,
            x='date_range',
            y='comb08',
            kind='box',
            col_wrap=2,
            showfliers=False,
           col='fuel_type_summary');
```

```python
sns.catplot(data=df, x="year", aspect=2, y="highway08", kind="point")
```

```python
sns.catplot(data=df, x="year", aspect=2, y="highway08", kind="point", hue="fuel_type_summary", join=False)
```

```python
sns.catplot(data=df, x="year", aspect=2, y="highway08", kind="point", hue="fuel_type_summary", linestyle="none")
```

```python
sns.catplot(data=df, x="year", aspect=2, y="highway08", kind="point", hue="fuel_type_summary")
```

```python
sns.catplot(data=df, 
            x="year",
            row="date_range",
            aspect=2, 
            y="highway08", 
            kind="bar", 
            hue="fuel_type_summary",
           sharex=False)
```

```python
sns.catplot(data=df, 
            x="cylinders",
            y="comb08", 
            row="date_range",
            col="drive",
            hue="transmission",
            kind="bar")
```
